```python
from larlib import *
```

```python
#definiamo una funzione facade che prende solo due numeri interi positivi
#restituisce un immagine simile ad un muro con il corrispettivo scheletro
def facade(n,m) :
    """Return an HPC object representing a facade of n columns and m rows."""
    xRithm = [5,-2,-5,-2]* n
    xRithm.append(5)
    yRithm = [7,-5,-2] * m 
    yRithm.append(7)
    xVoid=[]
    yVoid=[]
    for x in range (0,len(xRithm),1):
        xVoid.append(-xRithm[x])
    for y in range (0,len(yRithm),1):
        yVoid.append(-yRithm[y])
    cart=PROD([QUOTE(xRithm),QUOTE(yRithm)])
    cart1=PROD([QUOTE(xVoid),QUOTE(yRithm)])
    cart2=PROD([QUOTE(xRithm),QUOTE(yVoid)])
    skel_1=SKEL_1(PROD([QUOTE(xVoid),QUOTE(yVoid)]))
    results = STRUCT([cart,cart1,cart2,skel_1])
    return results
    
```

```python
VIEW(facade(4,6))
```
