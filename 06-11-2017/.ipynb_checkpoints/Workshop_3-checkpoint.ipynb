Leonardo Church
========

schema originale della chiesa di leonardo


```python
from pyplasm import *
from larlib import *
```

```python
def mezzaSfera(radius):
    def GSPHERE0(subds):
        N, M = subds
        domain = Hpc(Grid([N * [PI / N], M * [PI / M]]))
        domain = MAT([[1, 0, 0, 0], [-PI * 2, 1, 0, 0], [-PI, 0, 1, 0], [0, 0, 0, 1]])(domain)
        fx = lambda p: radius * math.cos(p[0]) * math.sin(p[1])
        fy = lambda p: radius * math.cos(p[0]) * math.cos(p[1])
        fz = lambda p: radius * math.sin(p[0])
        ret = GMAP([fx, fy, fz])(domain)
        return R([1, 2])(PI)(ret)
    return GSPHERE0
```

```python
def bigDome(radius):
    #generate the main dome with @parameter:radius

    def point(r, h):
        hFraction = .05 * h;
        base = CYLINDER([1.8 * r, 2 * hFraction])(30)
        lowerBlock = TOP([base, CUBOID([1.8 * r, 1.8 * r, hFraction])])
        middleBlock = TOP([lowerBlock, CUBOID([1.5 * r, 1.5 * r, 6. * hFraction])])
        block = TOP([middleBlock, CUBOID([1.8 * r, 1.8 * r, hFraction])])
        base2 = CYLINDER([.7 * r, 2. * hFraction])(30)
        top = TOP([base2, CONE([.7 * r, 7. * hFraction])(30)])
        return TOP([block,top])

    sphereLevel = .94 * radius
    layer1 =COLOR(BROWN) (CYLINDER([1.05 * radius, .1 * radius])(100))
    layer2 =COLOR(BROWN) (CYLINDER([1.03 * radius, .04 * radius])(100))

    #components
    base =COLOR(BROWN) (CYLINDER([1.05 * radius, .2 * radius])(100))
    sphere = COLOR(RED) (JOIN(mezzaSfera(radius)([30, 30])))
    pillar = COLOR(WHITE)(CYLINDER([radius, radius / 2])(100))
    point = COLOR(BROWN) (point(.1 * radius, .6 * radius))
    

    #composition
    dome = TOP([base, pillar])
    dome = TOP([dome, sphere])
    return dome


VIEW(bigDome(10))
```

```python
def smallDome(radius, height):
   #generate the model for the smallers domes @param: radius @param: height of pillar
    
    def point(r, h):
        return CONE([r, h])(30)
    
    arcsZ = 1.45 * radius
    pointZ = arcsZ + 1.05 * radius
    layer1 = CYLINDER([1.07 * radius, .05 * radius])(100)
    layer2 = CYLINDER([1.03 * radius, .1 * radius])(100)
    sphere = COLOR(RED)(JOIN(mezzaSfera(radius)([30,30])))
    pointZT = 1.35 * radius + height
    
    #components
    base =COLOR(BROWN) (CYLINDER([1.05 * radius, .2 * radius])(100))
    pillar =COLOR(WHITE) (CYLINDER([radius, height])(100))
    point = COLOR(BROWN)(point(.1 * radius, .2 * radius))
    
    #composition
    dome = TOP([base, pillar])
    dome = TOP([dome,sphere])
    return dome
    
VIEW(smallDome(4,4))
```

```python
def Building(x, y, z):
    #generate the main building with three parameter: with,depth and height
    
    lowerLevel =COLOR(BROWN) (CUBOID([.97 * x, .97 * y, .13 * z]))
    middleLevel =COLOR(WHITE)(CUBOID([.94 * x, .94 * y, .8 * z]))
    upperLevel =COLOR(BROWN) (CUBOID([x, y, .07 * z]))
    
    return TOP([ TOP([lowerLevel, middleLevel]) , upperLevel])

VIEW(Building(30, 30, 20))
```

```python
def externalStructure(x, y, z):
   #generate the model for the external structure for the main building with three parameter: with,depth and height
    
    dx = x / 8.
    dy = y / 8.
    dz = z / 7.
    domeZ = 2.5 * dz - 2.35 * dx
    pillarRadius = 1.8 * dx
    baseH = (.13 * 4.5 * dz) - (.2 * (pillarRadius))
    pillarBase = (CYLINDER([1.05 * (pillarRadius), baseH])(100))
    pillarDome = smallDome(pillarRadius, 1.5 * dz)
    domeBase =(CUBOID([4.5 * dx, 4.5 * dy, .2 * dz]))
    centralDome = smallDome(1.3 * dz, domeZ)
    
    #components
    centralBlock = Building(6 * dx , 6 * dy, 4 * dz)
    dome = TOP([domeBase, centralDome])
    pillar = TOP([ pillarBase, pillarDome ])
    
    #composition
    structure = TOP([centralBlock, dome])
    structure = STRUCT([structure,T([1])([3 * dx])(pillar),T([1,2])([3 * dx, 5.65 * dy])(pillar),T([1,2])([5.8 * dx, 3 * dy])(pillar)])
    return structure

VIEW(externalStructure(10, 10, 15))
```

```python
def ggpl_leonardoChurch(x, y, z):
    #generate the Leonardo Church
    
    #ratios
    ratioX = x / 80.
    ratioY = y / 80.
    ratioZ = z / 40.
    #components
    bDome = bigDome(10)
    building = Building(40, 40, 20)
    dome = smallDome(4.5, 3.5)
    extStr = externalStructure(20, 20, 26)
   
    
    #composition
    church = TOP([building, bDome])
    church = STRUCT([church,T([1,2,3])([7, 7, 20])(dome),T([1,2,3])([7, 32, 20])(dome),T([1,2,3])([32, 7, 20])(dome),T([1,2,3])([32, 32, 20])(dome),])
    church = STRUCT([church, \
                T([1,2])([31, 12])(extStr),\
                R([1,2])(PI / 2)(T([1,2])([31, -27])(extStr)),\
                R([1,2])(PI)(T([1,2])([-8, -27])(extStr)),\
                R([1,2])(3 * PI / 2), T([1,2])([-8, 12])(extStr),\
                ])
    return S([1,2,3])([ratioX, ratioY, ratioZ])(church)

VIEW(ggpl_leonardoChurch(80, 80, 40))
```
